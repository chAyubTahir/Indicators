```python
import pandas as pd
import numpy as np

data= pd.read_csv("D:\Deformed data.csv")

d = data.values
```

```python
window_size = 2
mse = np.zeros((d.shape[0]-window_size+1))
mase = np.zeros((d.shape[0]-window_size+1))
rmse = np.zeros((d.shape[0]-window_size+1))
standard_deviation = np.zeros((d.shape[0]-window_size+1))
median = np.zeros((d.shape[0]-window_size+1))

for i in range(d.shape[0]-window_size+1):
    print(i)
    diff = d[i:i+window_size, 0]
    median_diff = np.median(diff)
    st_d = np.std(diff)
    sq = np.square(diff)
    sm = np.sum(sq)
    sum_absolute = np.sum(np.abs(diff))
    mse[i] = sm / window_size
    root = np.sqrt(mse[i])  
    rmse[i] = root
    mase[i] = sum_absolute / window_size
    standard_deviation[i] = st_d 
    median[i] = median_diff
    print(mse[i])
    print(mase[i])
    print(rmse[i])
    print(standard_deviation[i])
    print(median[i])
```

```python
col = ['Mean_Squared_Error', 'Mean_Absolute_Error', 'Root_Mean_Squared_Error', 'Standard_Deviation'
      ,'Median']
```

```python
df_ = pd.DataFrame( columns=col)
```

```python
df_.Mean_Squared_Error = mse
df_.Mean_Absolute_Error = mase
df_.Root_Mean_Squared_Error = rmse
df_.Standard_Deviation = standard_deviation
df_.Median = median
```

```python
df_.to_csv('Actual_Indicators.csv', index=False)
```

```python
window_size = 2
mse_f = np.zeros((d.shape[0]-window_size+1))
mase_f = np.zeros((d.shape[0]-window_size+1))
rmse_f = np.zeros((d.shape[0]-window_size+1))
standard_deviation_f = np.zeros((d.shape[0]-window_size+1))
median_f = np.zeros((d.shape[0]-window_size+1))

for i in range(d.shape[0]-window_size+1):
    print(i)
    diff_f = d[i:i+window_size, 1]
    median_diff_f = np.median(diff_f)
    st_d_f = np.std(diff_f)
    sq_f = np.square(diff_f)
    sm_f = np.sum(sq_f)
    sum_absolute_f = np.sum(np.abs(diff_f))
    mse_f[i] = sm_f / window_size
    root_f = np.sqrt(mse_f[i])  
    rmse_f[i] = root_f
    mase_f[i] = sum_absolute_f / window_size
    standard_deviation_f[i] = st_d_f
    median_f[i] = median_diff_f
    print(mse_f[i])
    print(mase_f[i])
    print(rmse_f[i])
    print(standard_deviation_f[i])
    print(median_f[i])
```
